# limestone 과 soxdiff 시각화

**limestone 값이 증가함에 따라 soxdiff (절감정도) 가 증가하는 경향이 있음**

```python
# NaN 값을 평균값으로 대체
df = df.fillna(df.mean())

# 'SoxDiff_Unit3' 열 생성
df['SoxDiff_Unit3'] = df['InletSox_Unit3'] - df['OutletSox_Unit3']

# 3호기 Limestone 사용량과 SoxDiff 값 시각화
plt.figure(figsize=(10, 6))
plt.scatter(df['Limestone_Unit3'], df['SoxDiff_Unit3'], alpha=0.5)
plt.title('Limestone Usage vs. SoxDiff for Unit 3')
plt.xlabel('Limestone Usage (Unit 3)')
plt.ylabel('SoxDiff (Unit 3)')
plt.grid(True)
plt.show()
```

# limestone 사용량에 따른 OutletSox / InletSox 비율 시각화

**특별한 상관관계를 확인할 수 없다.**

```python
# InletSox > OutletSox인 경우만 필터링
df = df[df['InletSox_Unit3'] > df['OutletSox_Unit3']]

# OutletSox가 InletSox에 차지하는 비율 계산
df['Sox_Ratio_Unit3'] = df['OutletSox_Unit3'] / df['InletSox_Unit3'] * 100



# 3호기 Limestone 사용량에 따른 OutletSox / InletSox 비율 시각화
plt.figure(figsize=(10, 6))
plt.scatter(df['Limestone_Unit3'], df['Sox_Ratio_Unit3'], alpha=0.5)
plt.title('Limestone Usage vs. Outlet Sox / Inlet Sox Ratio for Unit 3')
plt.xlabel('Limestone Usage (Unit 3)')
plt.ylabel('Outlet Sox / Inlet Sox Ratio (%)')
plt.grid(True)
plt.show()
```
